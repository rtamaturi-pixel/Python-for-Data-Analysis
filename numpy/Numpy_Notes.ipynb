SPECIAL ARRAYS:

```python
import numpy as np
arr0 = np.zeros((3,5), dtype='int')
print(arr0)
```

```python
arr1s = np.ones((2,8), dtype='float')
print(arr1s)
```

```python
arrfull = np.full((4,4),7)    #an array with 4x4 shape and all values are 7
print(arrfull)
```

```python
arrId = np.eye(4)   #4x4 identity matrix
print(arrId)
```

```python
arrSeq = np.arange(5)  # array with values from 0 to 4
arrSeq
```

```python
arrSeq2 = np.arange(2,10,2)  # array with values from 2 to 9 with step size of 2
arrSeq2
```
