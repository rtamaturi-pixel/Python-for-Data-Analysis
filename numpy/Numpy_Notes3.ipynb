## Array Operations and Broadcasting:

```python
import numpy as np 
```

```python
a1 = np.array([1,2,3])
a2 = np.array([4,5,6])
```

```python
print(a1 + a2)  # Element-wise addition
print(a1 - a2)  # Element-wise subtraction
print(a1 * a2)  # Element-wise multiplication
print(a1 / a2)  # Element-wise division
```

```python
a1 + 10 # Broadcasting: add 10 to each element of a1
```

```python
a2 - 1 # Broadcasting: subtract 1 from each element of a2
```

```python
a2 * 3 # Broadcasting: multiply each element of a2 by 3
```

```python
arr2 = np.array([1,4,9,16])
print(np.sqrt(arr2))  # Element-wise square root
print(np.log(arr2))   # Element-wise natural logarithm
print(np.exp(arr2))   # Element-wise exponential
print(np.sin(arr2))   # Element-wise sine
print(np.cos(arr2))  # Element-wise cosine

```

```python
arr4 = np.array([[10,30,50],[25,66,34]])
np.sum(arr4)
```

```python
np.sum(arr4, axis=0)  # Sum along columns
```

```python
np.sum(arr4, axis=1)  # Sum along rows
```

```python
np.mean(arr4)  # Mean of all elements
np.max(arr4)   # Maximum element
np.min(arr4)   # Minimum element
np.std(arr4)   # Standard deviation of all elements
```

```python
arr5 = np.array([1,3,5],
                [3,5,6])
arr6 = np.array([7,8,9])
```

```python
np.sum(arr5,arr6)  # Element-wise addition of arr5 and arr6
```
