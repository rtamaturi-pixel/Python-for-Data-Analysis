## More on Data Frames:

```python
import numpy as np
import pandas as pd
```

### Creating a Random Data:

```python
random_data = np.random.randint(1,101,size=(3,5))  # r,c
random_data
# Creating a random data with 3 rows and 5 columns with values between 1 to 100.
```

### Converting Random data to Data Frame:

```python
row_labels = ["Morning","Afternoon","Evening"]
```

```python
column_labels = ["Monday","Tuesday","Wednesday","Thursday","Friday"]
```

```python
df = pd.DataFrame(data=random_data, index = row_labels, columns= column_labels)
```

```python
df
```

### Transposing the Data frame:

```python
df.T   # .T is used to Transpose the Data frame
```

```python
df.transpose()  # Another way to Transpose the Data frame
```

- axis = 1 : is for sorting Columns.

```python
nba.sort_index()  # ascending by default A-Z
```

```python
nba.sort_index(axis=1, ascending = False)  #axis=1 or axis = "columns" for sorting columns
```
