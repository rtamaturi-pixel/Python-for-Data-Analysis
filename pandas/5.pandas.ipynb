### isin() nethod:

```python
star_teams = ["Legal", "Sales", "Marketing"]
all_star_teams = employees["Team"].isin(star_teams)
employees[all_star_teams]
```

```python
salary_80k_90k = employees["Salary"].between(80000,90000)
```

```python
employees[salary_80k_90k]
```

```python
between_80_90 = employees["Start Date"].between(
    left="1980-01-01",
    right="1990-01-01"
)
employees[between_80_90]
```

### isnull() ---- Identifies null values in column

```python
employees["First Name"].isnull()
employees[employees["First Name"].isnull()]
```

```python
employees["Mgmt"].isnull()
employees[employees["Mgmt"].isnull()]
```

```python
# so here by the data, we undestrand the nulls in First Name are nulls in Mgmt column as well.
```

### dropna() ---- Drops rows with null values in specified columnm

```python
employees.dropna() # drop all the row missing values
```

```python
employees.dropna(how= "any")   # if atleast one null value in a row, such rows will be dropped.
```

```python
employees.dropna(how = "all")  # only if all values in a row are null, such rows will be dropped.
```

```python
employees.dropna(subset=["Gender"])  # drop rows where Gender is null.
```

```python
# change salary datatype from float to integer
employees["Salary"] = employees["Salary"].astype("Int64")
```

```python
employees.dropna(subset=["Gender", "First Name"])  # drop rows where Gender and First name is null.
```

### thresh specifies non - missing values to keep a row from deleting

```python
employees.dropna(thresh= 4) # drop rows which have less than 4 non null values.
```

### Inversion Operator: (~)
#### Inversion operator (~) helps to select non null values in a column and filter the dataframe based on that and convert the True to False and vice versa.

### Duplicated()

```python
employees["Team"].head(15).duplicated()   # check for duplicated values in Team column for first 15 rows.
```

```python
employees["Team"].duplicated(keep= "first")  # mark duplicates as True except for first occurrence.
```

```python
employees["Team"].duplicated(keep= "last")  # mark duplicates as True except for the last occurrence.
```
