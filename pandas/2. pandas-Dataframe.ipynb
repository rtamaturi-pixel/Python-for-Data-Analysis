## Let's Focus of "Gross" column:

```python
movies.Gross.dtype   

# 'O' means Object.
```

```python
movies.Gross.head(3)  

#since there is $ sign it became object type instead of numeric type.
```

```python
# now we will convert it to float type by removing $ sign and commas
```

```python
movies["Gross"].str  #string functions
```

```python
movies["Gross"].str.replace("$","").str.replace(",","")
```

### We have replaced the $ and comma(,) with empty, still it is Object DType.  so we need to convert its Data type to Float

```python
# for that we will use astype() function to convert the Data type to float
```

```python
movies["Gross"].str.replace("$","").str.replace(",","").astype(float)
```

### It is just a reference, to make the changes permanent we need to assign it back to the Gross column

```python
movies["Gross"] 

#it is still object type because we didn't assign the converted values back to the DataFrame.
```

```python
movies["Gross"] = (
    movies["Gross"]
    .str.replace("$", "")
    .str.replace(",", "")
    .astype(float)
)

# we assigned the converted values back to the DataFrame.
```

```python
movies["Gross"]   

# now it is float type
```

### sorting the columns.

```python
movies.sort_values(by="Year")  

# by default it sorts by ascending order
```

#### Sorting the columns by Multiple columns.

```python
movies.sort_values(by = ["Studio", "Year"])  

# sorting by multiple columns, here position of ordering is important. 
# first Studio will be sorted, and with respected to studio ,year will be sorted.     
```

#### if wanted descending order we can use ascending = False parameter.

```python
movies.sort_values(by = ["Studio", "Year"], ascending= [False, True])

# Studio will be Descending order, (asc = false)
# Year will be Ascending order, (asc = True)
```
