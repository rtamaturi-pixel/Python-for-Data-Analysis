## Multi Index Data Frame:

```python
import pandas as pd
import numpy as np
```

```python
columns = ["Morning", "Evening"]
rows = [" Mon", "Tue", "Wed"]
data = [
    (1,2),
    (3,4),
    (5,6)
]
pd.DataFrame(data= data, index=rows, columns=columns)
```

```python
column_index = pd.MultiIndex.from_tuples([
                ("AP", "Schools"),
                ("AP", "Police"),
                ("TS", "Schools"),
                ("TS", "Police")
])
rows = [" Mon", "Tue", "Wed"]
data = [
    (1,2),
    (3,4),
    (5,6)
]
pd.DataFrame(data= data, index=rows, columns=columns)
```
